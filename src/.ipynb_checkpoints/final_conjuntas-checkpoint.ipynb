## Energías

```julia
function get_vgv(id)
    modes = Symbol("modes_", id)
    evals = Symbol("evals_", id)
    vol_avg = Symbol("vol_avg_", id)
    vol_ndd = Symbol("vol_ndd_", id)
    vgv = Symbol("vgv_", id)
    
    # Leo modos
    @eval $modes, $evals = JUMD.readPtrajModes(joinpath(home, "1ure",  "pca",
        string("full_modes_", "1ure")))
    # Leo volumen original
    @eval $vol_avg = convert(Float64, readdlm(joinpath(home, "1ure",  "ndd",
        string("vol_avg_", "1ure")))[3])
    # Leo volúmenes NDD
    @eval $vol_ndd = convert(Array{Float64, 1},
        readdlm(joinpath(home, "1ure",  "ndd", string("out_ndd_", "1ure")))[2:end, 2])
    # VGV
    @eval ($vgv = ($vol_ndd .- $vol_avg) .* $evals)
    @eval ($vgv = $vgv ./ norm($vgv))
end

get_vgv("1ure");
get_vgv("2w9y");
get_vgv("4uet");
get_vgv("4xcp");
get_vgv("1ifb");
get_vgv("2ifb");
```

```julia
# Leo modos
modes_1ure, evals_1ure = JUMD.readPtrajModes(joinpath(home, "1ure",  "pca",
        string("full_modes_", "1ure")))
# Leo volumen original
vol_avg_1ure = convert(Float64, readdlm(joinpath(home, "1ure",  "ndd",
            string("vol_avg_", "1ure")))[3])
# Leo volúmenes NDD
ndd_out_1ure = convert(Array{Float64, 1},
    readdlm(joinpath(home, "1ure",  "ndd", string("out_ndd_", "1ure")))[2:end, 2])
# VGV
vgv_1ure = (ndd_out_1ure .- vol_avg_1ure) .* evals_1ure
vgv_1ure = vgv_1ure ./ norm(vgv_1ure);


# Leo modos
modes_2w9y, evals_2w9y = JUMD.readPtrajModes(joinpath(home, "2w9y",  "pca",
        string("full_modes_", "2w9y")))
# Leo volumen original
vol_avg_2w9y = convert(Float64, readdlm(joinpath(home, "2w9y",  "ndd",
            string("vol_avg_", "2w9y")))[3])
# Leo volúmenes NDD
ndd_out_2w9y = convert(Array{Float64, 1},
    readdlm(joinpath(home, "2w9y",  "ndd", string("out_ndd_", "2w9y")))[2:end, 2])
# VGV
vgv_2w9y = (ndd_out_2w9y .- vol_avg_2w9y) .* evals_2w9y
vgv_2w9y = vgv_2w9y ./ norm(vgv_2w9y);


# Leo modos
modes_4uet, evals_4uet = JUMD.readPtrajModes(joinpath(home, "4uet",  "pca",
        string("full_modes_", "4uet")))
# Leo volumen original
vol_avg_4uet = convert(Float64, readdlm(joinpath(home, "4uet",  "ndd",
            string("vol_avg_", "4uet")))[3])
# Leo volúmenes NDD
ndd_out_4uet = convert(Array{Float64, 1},
    readdlm(joinpath(home, "4uet",  "ndd", string("out_ndd_", "4uet")))[2:end, 2])
# VGV
vgv_4uet = (ndd_out_4uet .- vol_avg_4uet) .* evals_4uet
vgv_4uet = vgv_4uet ./ norm(vgv_4uet);


# Leo modos
modes_4xcp, evals_4xcp = JUMD.readPtrajModes(
    joinpath(home, "4xcp",  "pca", string("full_modes_", "4xcp")))
# Leo volumen original
vol_avg_4xcp = convert(Float64, readdlm(joinpath(home, "4xcp",  "ndd",
            string("vol_avg_", "4xcp")))[3])
# Leo volúmenes NDD
ndd_out_4xcp = convert(Array{Float64, 1},
    readdlm(joinpath(home, "4xcp",  "ndd", string("out_ndd_", "4xcp")))[2:end, 2])

vgv_4xcp = (ndd_out_4xcp .- vol_avg_4xcp) .* evals_4xcp
vgv_4xcp = vgv_4xcp ./ norm(vgv_4xcp);

# Leo modos
modes_1ifb, evals_1ifb = JUMD.readPtrajModes(
    joinpath(home, "1ifb",  "pca", string("modes_", "1ifb")))
# Leo volumen original
vol_avg_1ifb = convert(Float64, readdlm(joinpath(home, "1ifb",  "ndd",
            string("vol_avg_", "1ifb")))[3])
ndd_out_1ifb = convert(Array{Float64, 1},
    readdlm(joinpath(home, "1ifb",  "ndd", string("out_ndd_", "1ifb")))[2:end, 2])
# VGV
vgv_1ifb = (ndd_out_1ifb .- vol_avg_1ifb) .* evals_1ifb
vgv_1ifb = vgv_1ifb ./ norm(vgv_1ifb);


# Leo modos
modes_2ifb, evals_2ifb = JUMD.readPtrajModes(
    joinpath(home, "2ifb",  "pca", string("modes_", "2ifb")))
# Leo volumen original
vol_avg_2ifb = convert(Float64, readdlm(joinpath(home, "2ifb",  "ndd",
            string("vol_avg_", "2ifb")))[3])
# Leo volúmenes NDD
ndd_out_2ifb = convert(Array{Float64, 1},
    readdlm(joinpath(home, "2ifb",  "ndd", string("out_ndd_", "2ifb")))[2:end, 2])
# VGV
vgv_2ifb = (ndd_out_2ifb .- vol_avg_2ifb) .* evals_2ifb
vgv_2ifb = vgv_2ifb ./ norm(vgv_2ifb);


# Leo modos
modes_1, evals_1 = JUMD.readPtrajModes(joinpath(home, "4xcp", "poblaciones",
        "pca", "1", string("1", "vol_modes_", "4xcp")))
# Leo volumen original
vol_avg_1 = convert(Float64, readdlm(joinpath(home, "4xcp", "poblaciones",
            "ndd", "1", string("vol_avg_", "4xcp")))[3])
# Leo volúmenes NDD
ndd_out_1 = convert(Array{Float64, 1},
    readdlm(joinpath(home, "4xcp", "poblaciones", "ndd", "1", string("out_ndd_", "4xcp")))[2:end, 2])
# VGV
vgv_1 = (ndd_out_1 .- vol_avg_1) .* evals_1
vgv_1 = vgv_1 ./ norm(vgv_1);

# Leo modos
modes_2, evals_2 = JUMD.readPtrajModes(joinpath(home, "4xcp", "poblaciones",
        "pca", "2", string("2", "vol_modes_", "4xcp")))
# Leo volumen original
vol_avg_2 = convert(Float64, readdlm(joinpath(home, "4xcp", "poblaciones",
            "ndd", "2", string("vol_avg_", "4xcp")))[3])
# Leo volúmenes NDD
ndd_out_2 = convert(Array{Float64, 1},
    readdlm(joinpath(home, "4xcp", "poblaciones", "ndd", "2", string("out_ndd_", "4xcp")))[2:end, 2])
# VGV
vgv_2 = (ndd_out_2 .- vol_avg_2) .* evals_2
vgv_2 = vgv_2 ./ norm(vgv_2);

# Leo modos
modes_3, evals_3 = JUMD.readPtrajModes(joinpath(home, "4xcp", "poblaciones",
        "pca", "3", string("3", "vol_modes_", "4xcp")))
# Leo volumen original
vol_avg_3 = convert(Float64, readdlm(joinpath(home, "4xcp", "poblaciones",
            "ndd", "3", string("vol_avg_", "4xcp")))[3])
# Leo volúmenes NDD
ndd_out_3 = convert(Array{Float64, 1},
    readdlm(joinpath(home, "4xcp", "poblaciones", "ndd", "3", string("out_ndd_", "4xcp")))[2:end, 2])
# VGV
vgv_3 = (ndd_out_3 .- vol_avg_3) .* evals_3
vgv_3 = vgv_3 ./ norm(vgv_3);
```

```julia
pasos = collect(-.8:.05:.8)
npasos = length(pasos)
RT = 0.693

E_1ure = Array{Float64,1}(undef, npasos)
[ E_1ure[i] = JUMD.energiaGdte(evals_1ure, vgv_1ure, pasos[i]) for i in 1:npasos ]
        
E_2w9y = Array{Float64,1}(undef, npasos)
[ E_2w9y[i] = JUMD.energiaGdte(evals_2w9y, vgv_2w9y, pasos[i]) for i in 1:npasos ]

E_4uet = Array{Float64,1}(undef, npasos)
[ E_4uet[i] = JUMD.energiaGdte(evals_4uet, vgv_4uet, pasos[i]) for i in 1:npasos ]

E_4xcp = Array{Float64,1}(undef, npasos)
[ E_4xcp[i] = JUMD.energiaGdte(evals_4xcp, vgv_4xcp, pasos[i]) for i in 1:npasos ]

E_1 = Array{Float64, 1}(undef, npasos)
[ E_1[i] = JUMD.energiaGdte(evals_1, vgv_1, pasos[i]) for i in 1:npasos]

E_2 = Array{Float64, 1}(undef, npasos)
[ E_2[i] = JUMD.energiaGdte(evals_2, vgv_2, pasos[i]) for i in 1:npasos]

E_3 = Array{Float64, 1}(undef, npasos)
[ E_3[i] = JUMD.energiaGdte(evals_3, vgv_3, pasos[i]) for i in 1:npasos];
```

```julia
plt_u = plot(pasos, E_1ure, lab = "E_1ure",
    xticks = -.7:.2:.7, yticks = 0:.2:1.,   
    xlims = (-.8, .8), ylims = (0, 1.),
    guidefont = font(16, "Helvetica"), tickfont = font(11, "Helvetica"),
    yaxis = L"U $[Kcal]$", xaxis = L"Displacement $[\AA]$ ",
    linecolor = Colors.colorant"#008080", linewidth = 3,
    grid = false)

plot!(plt_u, pasos, E_2w9y, lab = "E_2w9y",
    linecolor = Colors.colorant"#32CD32", linewidth = 3)
plot!(plt_u, pasos, E_4uet, lab = "E_4uet",
    linecolor = Colors.colorant"#0000FF", linewidth = 3)

plot!(plt_u, pasos, E_4xcp, lab = "E_4xcp",
    linecolor = Colors.colorant"#FF0000", linewidth = 3)
plot!(plt_u, pasos, E_1, lab = "Conformer 1",
    linecolor = Colors.colorant"#d64161", linewidth = 3)
plot!(plt_u, pasos, E_2, lab = "Conformer 2",
    linecolor = Colors.colorant"#feb236", linewidth = 3)
plot!(plt_u, pasos, E_3, lab = "Conformer 3",
    linecolor = Colors.colorant"#6b5b95", linewidth = 3)
hline!(plt_u, [RT], lab = "RT", linecolor = Colors.colorant"black")
```

```julia
savefig(plt_u, joinpath(rtdos, "papel", "7", string("7_U_", ".svg")))
```

## Volumen

```julia
trj_vol_4uet = convert(Array{Float64, 1}, 
    readdlm(joinpath(home, "4uet", "cavidad", string("vol_trj_4uet")))[:, 3]);

trj_vol_4xcp = convert(Array{Float64, 1}, 
    readdlm(joinpath(home, "4xcp", "cavidad", string("vol_trj_4xcp")))[:, 3]);
st = 40
bin_vol = collect(700:st:2200)
wgh_vol = fill(1.0, length(trj_vol))

# Histograma ponderado
Wbins_vol_4uet, Whis_vol_4uet = JUMD.weightedHist(trj_vol_4uet, bin_vol, wgh_vol, true, false);
Wbins_vol_4xcp, Whis_vol_4xcp = JUMD.weightedHist(trj_vol_4xcp, bin_vol, wgh_vol, true, false);

grp = repeat(["4UET", "4XCP"], inner = length(Wbins_vol_4xcp))
clr = repeat([colorant"#0000FF", colorant"#FF0000"],
    inner = length(Wbins_vol_4xcp))

ambas_uet_xcp = groupedbar(
    repeat(Wbins_vol_4uet, outer = 2), hcat(Whis_vol_4uet, Whis_vol_4xcp),
    group = grp, fillcolor = clr,
    xticks = 1E3:250:2E3, yticks = .025:.025:.75,
    xlims = (700, 2300), ylims = (0, .075),
    
    linecolor = false, legend = (.9, .95),
    grid = false, guidefont = font(16, "Arial"), tickfont = font(11, "Arial"),
    yaxis = "Density", xaxis = L"Volume $[\AA^3]$ ")
```

```julia
savefig(ambas_uet_xcp, joinpath(rtdos, "papel", "1", string("ambas_uet_xcp.svg")))
```
